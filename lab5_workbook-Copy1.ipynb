```python
import itertools as it
import pandas as pd
import numpy as np

num = list(range(1,7))
roll = list(it.product(num,repeat=3))
print(f"There is a total of {len(roll)} possible outcomes")
s = np.array(roll)
sums = np.sum(s,axis=1)
df = pd.DataFrame({"Outcome":roll,"Outcome Sum":sums})
df
```

```python
x=df.groupby("Outcome Sum")
y= x.count()
z= y.values[:,0]
probability = z/ np.sum(z)
df1 = pd.DataFrame({"Sum":list(range(3,19)),"# of times outcome occurred":z,"Probability":probability})
df1
```
